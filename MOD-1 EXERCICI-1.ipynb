```python
"""
MODUL 1 EXERCICI 1
Crea una funció que donat un Array d’una dimensió, et faci un resum estadístic bàsic de les dades. Si detecta que 
l’array té més d’una dimensió, ha de mostrar un missatge d’error.
"""

import numpy as np
from numpy import random

##############
###FUNCIONS###
##############

#funció per obtenir el nombre total de nombres
def calcularElements(estructura):
    if (len(estructura) > 1):
        elements = estructura[0] * calcularElements(estructura[1:])
    else:
        elements = estructura[0]
    return elements

#funció per crear arrays
def crearMatriu(minim, maxim, elements, sonFlotants):
    rang = maxim - minim
    if (sonFlotants):
        matriu = minim + (rang * random.rand(elements))
    else:
        matriu = minim + random.randint(rang, size = elements)
    return matriu

#funció per mostrar les característiques de la matriu generada a l'atzar
def mostrarCaracteristiquesMatriu(matriu, index, dimensions, estructura, elements, minim, maxim, sonFlotants):
    mostrarTitul("INFORMACIO DE L'ARRAY Nº " + str(index+1) + " GENERADA A L'ATZAR")
    mostrarValor("Dimensions", dimensions)
    mostrarValor("Estructura", estructura)
    mostrarValor("Shape", matriu.shape)
    mostrarValor("Elements", elements)
    mostrarValor("Són Flotants", sonFlotants)
    mostrarValor("Límit minim", minim)
    mostrarValor("Límit maxim", maxim)
    mostrarValor("Matriu", matriu)

#funció per mostrar estadístiques bàsiques
def mostrarEstadistiquesMatriu(matriu):
    mostrarValor("mitja", np.mean(matriu))
    mostrarValor("mitjana", np.median(matriu))
    mostrarValor("màxim", np.max(matriu))
    mostrarValor("mímim", np.min(matriu))
    mostrarValor("Desviació estàndard", np.std(matriu))

#funció per mostrar títuls
def mostrarTitul(titul):
    MARGELATERAL = 10
    DECORACIO = "*"
    senefa = DECORACIO * (len(titul) + (MARGELATERAL * 2))
    senefaLateral = DECORACIO * MARGELATERAL
    text =  senefaLateral + titul + senefaLateral
    print(senefa)
    print(text)
    print(senefa)

#funció per mostrar un valor
def mostrarValor(text, valor):
    TABULACIO = 3
    tabulacio = " " * TABULACIO
    print(tabulacio + text + "=" + str(valor))    

#funció per mostrar una alerta
def mostrarAlerta(text):
    senefa = "-" * len(text)
    print(senefa)
    print(text)
    print(senefa)

###############
###CONSTANTS###
###############

REPETICIONS = 10
DIMENSIONS = [1, 2, 3]
PROBABILITATDIMENSIONS = [0.8, 0.1, 0.1]
MINIMDELEMENTS = 1
MAXIMDELEMENTS = 10
OPCIONSFLOTANTS = [False, True]
PROBABILITATFLOTANTS = [0.8, 0.2]
LIMIT = 1000
ZERO = 0
DIMENSIONSPERMESES = 1
ALERTAKO = "Aquesta array te més d'una dimensió. No es pot continuar."
ALERTAOK = "Aquesta array te una única dimensió, per tant, fem estadístiques:"

##############
###PROGRAMA###
##############

for index in range(REPETICIONS):
    dimensions = random.choice(DIMENSIONS, p = PROBABILITATDIMENSIONS)
    #estructura = MINIMDELEMENTS + random.randint(MAXIMDELEMENTS, size=dimensions)
    estructura = crearMatriu(MINIMDELEMENTS, MAXIMDELEMENTS, dimensions, False)
    elements = calcularElements(estructura)
    sonFlotants = random.choice(OPCIONSFLOTANTS, p = PROBABILITATFLOTANTS)
    
    atzar = random.randint(LIMIT) - random.randint(LIMIT) 
    if (atzar > ZERO): 
        maxim = atzar
        minim = maxim - random.randint(LIMIT)
    else:
        minim = atzar
        maxim = minim + random.randint(LIMIT)

    #creo una primera matriu de 1-D amb tots els elements
    matriu = crearMatriu(minim, maxim, elements, sonFlotants)
    #reshape la matriu amb l'estructura escollida
    matriu = matriu.reshape(estructura)
    mostrarCaracteristiquesMatriu(matriu, index, dimensions, estructura, elements, minim, maxim, sonFlotants)

    if (len(matriu.shape) > DIMENSIONSPERMESES):
        mostrarAlerta(ALERTAKO)
    else:
        mostrarAlerta(ALERTAOK)
        mostrarEstadistiquesMatriu(matriu)
```
