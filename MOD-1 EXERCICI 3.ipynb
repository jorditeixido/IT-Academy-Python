```python
"""
MODUL 1 EXERCICI 3
Crea una funció que donada una taula de dues dimensions, et calculi els totals per fila i els totals per columna.
"""

import numpy as np
from numpy import random

import statistics

##############
###FUNCIONS###
##############

#funció per mostrar títuls
def mostrarTitul(titul):
    MARGELATERAL = 10
    DECORACIO = "*"
    senefa = DECORACIO * (len(titul) + (MARGELATERAL * 2))
    senefaLateral = DECORACIO * MARGELATERAL
    text =  senefaLateral + titul + senefaLateral
    print(senefa)
    print(text)
    print(senefa)

#funció que imprimeix la suma de cada linia
def sumarLinies(matriu):
    for linia in range(matriu.shape[0]): 
        suma = sumarLinia(matriu,linia)
        mostrarSuma(LINIA, linia, suma)

#funció per sumar el total d'una linia
def sumarLinia(matriu, linia):
    return sum(matriu[linia])

#funció que imprimeix la suma de cada columna
def sumarColumnes(matriu):
    for columna in range(matriu.shape[1]):
        suma = sumarColumna(matriu, columna)
        mostrarSuma(COLUMNA, columna, suma)
    
#funció suma el total d'una columna
def sumarColumna(matriu, columna):
    suma=0;
    for linia in range(matriu.shape[0]):
        suma += matriu[linia][columna]
    return suma

#funció per mostrar resultats
def mostrarSuma(literal, index, suma):
    print (literal + str(index) + IGUAL + str(suma))

###############
###CONSTANTS###
###############

MINIM = 0 
MAXIM = 100 
NELEMENTS = 10 
LINIA = "Línia " 
COLUMNA = "Columna " 
IGUAL = " = " 
REPETICIONS = 10
MATRIUGENERADA = "MATRIU GENERADA "
SUMALINIES = "SUMA DE LÍNIES"
SUMACOLUMNES = "SUMA DE COLUMNES"

##############
###PROGRAMA###
##############

for index in range(REPETICIONS):

    nLinies = random.randint(3, 10)
    nColumnes = random.randint(3, 10)

    matriu = MINIM + (random.randint(MAXIM, size = nLinies * nColumnes))
    matriu = matriu.reshape(nLinies, nColumnes)

    mostrarTitul(MATRIUGENERADA + str(index))
    print(matriu)

    mostrarTitul(SUMALINIES)
    sumarLinies(matriu)

    mostrarTitul(SUMACOLUMNES)
    sumarColumnes(matriu)
```
